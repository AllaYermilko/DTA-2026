### Завдання 17. Калькулятор ⭐⭐

Запитайте у користувача два числа і знак операції (`+`, `-`, `*`, `/`).
Виведіть результат відповідної операції.

**Приклад:**
```
Перше число: 10
Друге число: 3
Знак: *
Результат: 30
```

⚠️ Не забудьте перевірити ділення на 0!

````python
# Ваш код для завдання 17:
a = float(input("Перше число: "))
b = float(input("Друге число: "))
sign = input("Знак: ")

if sign == "+":
    result = a + b
elif sign == "-":
    result = a - b
elif sign == "*":
    result = a * b
elif sign == "/":
    if b != 0:
        result = a / b
    else:
        result = "Помилка (ділення на нуль)"
else:
    result = "Невідомий знак"

print("Результат:", result)
````

### Завдання 27. Прості числа ⭐⭐⭐

Виведіть всі прості числа від 2 до 50.

**Нагадування:** просте число — це число, яке ділиться тільки на 1 і саме на себе.

> 💡 **Підказка:** Для кожного числа `n` від 2 до 50 перевіряйте, чи має воно дільники від 2 до n-1. Можна використати змінну-прапорець `is_prime = True`.

````python
# Ваш код для завдання 27:
for n in range(2, 51):
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(n)
````

### Завдання 31. Підрахунок голосних ⭐⭐

> 💡 **Підказка:** Створіть лічильник `count = 0`. У циклі перевіряйте: `if letter in "аеиіоуяюєї":` і збільшуйте лічильник.

**Алгоритм:** доки `b ≠ 0`: одночасно `a, b = b, a % b`. Коли `b = 0`, відповідь — це `a`.

**Приклад:** НСД(48, 36) = 12.

> 💡 **Підказка:** Найпростіший алгоритм — «бульбашкове сортування»: у вкладеному циклі порівнюємо сусідні елементи і міняємо їх місцями, якщо ліворуч стоїть більший. Повторюємо, поки не пройдемо весь список без жодної заміни.

````python
# Ваш код для завдання 56:
numbers = [42, 17, 89, 3, 56, 91, 24]
n = len(numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]

print(numbers)
````

- Підрахуйте кількість спроб і виведіть її в кінці

````python
# Ваш код для завдання 59:
secret = 42
attempts = 0
guessed = False

while not guessed:
    user_num = int(input("Вгадайте число: "))
    attempts += 1

    if user_num < secret:
        print("Більше")
    elif user_num > secret:
        print("Менше")
    else:
        print("Вітаю!")
        guessed = True

print(f"Ви вгадали число за {attempts} спроб(и).")
````

````python
# Ваш код для завдання 60:
sentence = "програмування на python це цікаво і корисно"

words = sentence.split()
word_count = len(words)

letters_only = ""
vowels_count = 0
vowels = "аеєиіїоуюяАЕЄИІЇОУЮЯaeiouyAEIOUY"

for char in sentence:
    if char.isalpha():
        letters_only += char
        if char in vowels:
            vowels_count += 1

longest_word = ""
for word in words:
    clean_word = word.strip(".,!—")
    if len(clean_word) > len(longest_word):
        longest_word = clean_word

print(f"Кількість літер (без пробілів): {len(letters_only)}")
print(f"Кількість слів: {word_count}")
print(f"Кількість голосних літер: {vowels_count}")
print(f"Найдовше слово: {longest_word}")
````

3. Найкращий день (з найбільшою виручкою) — назвою (Понеділок, Вівторок...)
4. Кількість днів, коли виручка була вищою за середню

> 💡 **Підказка:** Створіть список з назвами днів: `days = ["Понеділок", "Вівторок", ...]`. Індекс найкращого дня знайдіть у циклі (або через `sales.index(max(sales))`).

### Завдання 62. Турнірна таблиця ⭐⭐⭐

Дано два списки: команди та їхні очки (відповідні позиції).
Виведіть турнірну таблицю, ВІДСОРТОВАНУ за спаданням очок, у такому форматі:
```
1. Шахтар — 67 очок
2. Динамо — 64 очки
...
```

> 💡 **Підказка:** Найпростіший спосіб: у циклі знаходьте команду з максимальними очками, виводьте її, потім видаляйте цю пару (команду і очки) з обох списків. Повторюйте, поки списки не спорожніють.

````python
teams = ["Динамо", "Шахтар", "Зоря", "Дніпро", "Ворскла", "Олександрія"]
points = [64, 67, 52, 48, 39, 45]
# Ваш код:
teams_copy = list(teams)
points_copy = list(points)

rank = 1

while points_copy:
    max_points = max(points_copy)
    max_index = points_copy.index(max_points)
    best_team = teams_copy[max_index]
    print(f"{rank}. {best_team} — {max_points} очок")
    teams_copy.pop(max_index)
    points_copy.pop(max_index)
    rank += 1
````
