### ✏️ Завдання 1.1

У тебе є дані середньомісячних температур у Харкові за рік (°C):

```
-5, -4, 2, 10, 17, 21, 23, 22, 16, 9, 2, -3
```

1. Створи NumPy масив із цих значень і назви його `temps`
2. Виведи масив
3. Виведи: кількість вимірів (`ndim`), форму (`shape`), кількість елементів (`size`), тип даних (`dtype`)
4. Виведи мінімальну, максимальну температуру та середньорічну (округли до 1 знаку)

````python
import numpy as np

# Твій код:
# Створи NumPy масив із цих значень і назви його temps
# Виведи масив
temps = np.array([-5, -4, 2, 10, 17, 21, 23, 22, 16, 9, 2, -3])
print("Середньомісячні температури у Харкові:", temps, "\n")

# Виведи: кількість вимірів (ndim), форму (shape), кількість елементів (size), тип даних (dtype)
print(temps.ndim)
print(temps.shape)
print(temps.size)
print(temps.dtype,"\n")

# Виведи мінімальну, максимальну температуру та середньорічну (округли до 1 знаку)
print("Середньорічна температура:   ", temps.mean().round(1))
print("Мінімальна температура:      ", temps.min().round(1))
print("Максимальна температура:     ", temps.max().round(1))
````

### ✏️ Завдання 3.1

Є масив цін товарів у доларах. Виконай **без циклів**:

```python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])
```

1. Переведи ціни в гривні (курс: **1 USD = 41.5 UAH**), результат округли до 2 знаків
2. Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
3. Знайди: мінімальну, максимальну, середню ціну з ПДВ
4. Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
5. Виведи результати з поясненнями

````python
prices_usd = np.array([15.99, 42.50, 8.00, 125.00, 33.75, 67.20, 9.99])

# Твій код:
# Переведи ціни в гривні (курс: 1 USD = 41.5 UAH), результат округли до 2 знаків
prices_uah = prices_usd * 41.5
print("Ціни в гривнях:", prices_uah.round(2))

# Нарахуй ПДВ 20% — помнож гривневі ціни на 1.2, округли до 2 знаків
prices_vat_uah = prices_uah * 1.2
print("З ПДВ (20%):", prices_vat_uah.round(2), "\n")

# Знайди: мінімальну, максимальну, середню ціну з ПДВ
print("Середня ціна з ПДВ:          ", prices_vat_uah.mean().round(2), "грн")
print("Мінімальна ціна з ПДВ:       ", prices_vat_uah.min(), "грн")
print("Максимальна ціна з ПДВ:      ", prices_vat_uah.max(), "грн", "\n")

# Знайди суму всіх товарів з ПДВ — це загальна вартість кошика
print("Загальна вартість кошика:    ", prices_vat_uah.sum().round(2), "грн")
````

### ✏️ Завдання 3.2

Є дані про кількість товарів на складі та їх собівартість.
Виконай **без циклів**:

```python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']
```

1. Порахуй загальну вартість кожного товару на складі (`quantity * cost_each`)
2. Знайди загальну вартість всього складу
3. Знайди товар з найбільшою сумарною вартістю — використай `np.argmax()`
4. Знайди товар з найменшою сумарною вартістю — `np.argmin()`
5. Виведи середню вартість одного найменування на складі

````python
quantity  = np.array([120, 45, 300, 18, 75, 200, 60])
cost_each = np.array([250, 1800, 45, 9500, 380, 120, 750])
names     = ['Кабель', 'Клавіатура', 'Зошит', 'Монітор',
             'Мишка', 'Ручка', 'Навушники']

# Твій код:
# Порахуй загальну вартість кожного товару на складі (quantity * cost_each)
total_value = quantity * cost_each
print("Загальна вартість кожного товару:", total_value)

# Знайди загальну вартість всього складу
print("Загальна вартість всього складу:", total_value.sum(), "грн")

# Знайди товар з найбільшою сумарною вартістю — використай np.argmax()
best_product = np.argmax(total_value)
print(f"Товар з найбільшою сумарною вартістю: {names[best_product]} з сумою {total_value[best_product]} грн")

# Знайди товар з найменшою сумарною вартістю — np.argmin()
weak_product = np.argmin(total_value)
print(f"Товар з найменшою сумарною вартістю: {names[weak_product]} з сумою {total_value[weak_product]} грн")

# Виведи середню вартість одного найменування на складі
print("Середня вартість одного найменування на складі:", cost_each.mean().round(2), "грн")
````

### ✏️ Завдання 4.1

Є масив часу виконання замовлень у хвилинах (норма — до 30 хв):

```python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])
```

1. Знайди всі замовлення, виконані **вчасно** (≤ 30 хв)
2. Знайди всі **запізнілі** замовлення (> 30 хв)
3. Порахуй кількість вчасних і запізнілих
4. Знайди відсоток запізнілих замовлень (округли до 1 знаку)
5. Яка середня затримка серед запізнілих замовлень?

````python
delivery_time = np.array([22, 35, 28, 41, 19, 55, 27, 33, 24, 48,
                           31, 26, 52, 18, 29, 38, 23, 44, 20, 36])

# Твій код:
# Знайди всі замовлення, виконані вчасно (≤ 30 хв)
in_time = delivery_time[delivery_time <= 30]
print("Замовлення, виконані вчасно (≤ 30 хв):", in_time)

# Знайди всі запізнілі замовлення (> 30 хв)
delayed = delivery_time[delivery_time > 30]
print("Запізнілі замовлення (> 30 хв):", delayed)

# Порахуй кількість вчасних і запізнілих
print("кількість вчасних замовлень:   ", in_time.size)
print("кількість запізнілих замовлень:", delayed.size)

# Знайди відсоток запізнілих замовлень (округли до 1 знаку)
percent_delayed = round((delayed.size / delivery_time.size) * 100, 1)
print("Bідсоток запізнілих замовлень:", percent_delayed, "%")

# Яка середня затримка серед запізнілих замовлень?
avg_delay = delayed.mean()
print("Cередня затримка серед запізнілих замовлень:", avg_delay.round(), "хв")
````

### ✏️ Завдання 4.2

Є масив щоденних продажів за місяць. Використай `np.where()`:

```python
np.random.seed(42)
daily = np.random.randint(100, 900, size=30)
```

1. Знайди дні з продажами **вище середнього** за місяць
2. Порахуй скільки таких днів
3. За допомогою `np.where()` створи масив `status`:
   - `'Гарний день'` — якщо продажі вище середнього
   - `'Звичайний день'` — якщо нижче або рівно
4. Виведи перші 10 днів: продажі та їх статус поряд

````python
np.random.seed(42)
daily = np.random.randint(100, 900, size=30)

# Твій код:
# Знайди дні з продажами вище середнього за місяць
print("Щоденних продажів за місяць: ", daily, "\n")
good_sales = daily[daily > daily.mean()]
print("Дні з продажами вище середнього за місяць: ", good_sales)

# Порахуй скільки таких днів
print("Кількість днів з продажами вище середнього за місяць: ", good_sales.size, "\n")

# За допомогою np.where() створи масив status:
# 'Гарний день' — якщо продажі вище середнього
# 'Звичайний день' — якщо нижче або рівно

status = np.where(daily > daily.mean(), "Гарний день", "Звичайний день")
print(status, "\n")

# Виведи перші 10 днів: продажі та їх статус поряд
for i in range(10):
    print(f"День {i + 1}: Продажі = {daily[i]} -> {status[i]}")
````

### ✏️ Завдання 5.1

Є 1D масив з 24 показників продажів (по годинах за добу):

```python
hourly = np.array([12, 8, 5, 3, 4, 9, 18, 35, 52, 61, 58, 64,
                   70, 67, 55, 48, 53, 62, 74, 68, 45, 30, 22, 15])
```

1. Перетвори масив у матрицю **6×4** (6 блоків по 4 години) і виведи
2. Знайди суму продажів у кожному 4-годинному блоці (по рядках)
3. Відсортуй вихідний 1D масив за **зростанням** і виведи
4. Знайди **індекси** годин відсортованих за зростанням продажів — `np.argsort()`
5. Виведи топ-3 найпродуктивніших години (їхні номери, рахуючи з 0)

````python
hourly = np.array([12, 8, 5, 3, 4, 9, 18, 35, 52, 61, 58, 64,
                   70, 67, 55, 48, 53, 62, 74, 68, 45, 30, 22, 15])

# Твій код:
# Перетвори масив у матрицю 6×4 (6 блоків по 4 години) і виведи
matrix_6x4 = hourly.reshape(6, 4)
print("Matrix n6×4:\n", matrix_6x4, "\n")

# Знайди суму продажів у кожному 4-годинному блоці (по рядках).
print("Сума продажів у кожному 4-годинному блоці:", matrix_6x4.sum(axis=1), "\n")

# Відсортуй вихідний 1D масив за зростанням і виведи
print("Відсортований масив (↑):  ", np.sort(hourly), "\n")

# Знайди індекси годин відсортованих за зростанням продажів — np.argsort()
print("Індекси сортування годин: ", np.argsort(hourly), "\n")

# Виведи топ-3 найпродуктивніших години (їхні номери, рахуючи з 0)
top3 = np.sort(hourly)[-3:]
print("Топ-3:", top3, "\n")
````

## Розділ 6 — Генерація випадкових чисел

> **Що повторюємо:** `np.random.seed()`, `np.random.randint()`,
> `np.random.normal()`, `np.random.choice()`

### ✏️ Завдання 6.1

Змоделюй результати маркетингового опитування 50 клієнтів.

1. Встанови `np.random.seed(99)`
2. Згенеруй масив `ages` — вік 50 клієнтів: цілі числа від 18 до 65
3. Згенеруй масив `scores` — оцінки задоволеності (нормальний розподіл, середнє=7, σ=1.5), округли до 1 знаку
4. Обріж `scores` так щоб значення були від 1 до 10 — `np.clip(arr, 1, 10)`
5. Виведи:
   - середній вік клієнтів
   - середню оцінку задоволеності
   - кількість клієнтів з оцінкою >= 8 («задоволені»)
   - відсоток задоволених

````python
# Твій код:
# 1. Встанови np.random.seed(99)
np.random.seed(99)

# 2. Згенеруй масив ages — вік 50 клієнтів: цілі числа від 18 до 65
ages = np.random.randint(18, 66, size=50)
print("Вік клієнтів:\n",ages, "\n")

# 3. Згенеруй масив scores — оцінки задоволеності (нормальний розподіл, середнє=7, σ=1.5), округли до 1 знаку
scores = np.round(np.random.normal(loc=7, scale=1.5, size=50), 1)
print("Оцінки задоволеності:\n", scores, "\n")

# 4. Обріж scores так щоб значення були від 1 до 10 — np.clip(arr, 1, 10)
scores_clip = np.clip(scores, 1, 10)
print("Оцінки задоволеності від 1 до 10:\n",scores_clip, "\n")

# 5. Виведи:
# - середній вік клієнтів
mean_age = np.mean(ages)
print(f"• Середній вік клієнтів: {mean_age:.1f} років")

# - середню оцінку задоволеності
mean_score = np.mean(scores)
print(f"• Середня оцінка задоволеності: {mean_score:.2f}")

# - кількість клієнтів з оцінкою >= 8 («задоволені»)
satisfied_count = np.sum(scores >= 8)
print(f"• Кількість клієнтів з оцінкою >= 8 («задоволені»): {satisfied_count}")

# - відсоток задоволених
satisfied_percentage = (satisfied_count / 50) * 100
print(f"• Відсоток задоволених: {satisfied_percentage:.1f}%")
````

Тобі передали масив продажів за 8 тижнів (56 днів).

Виконай повний аналіз та дай відповіді на запитання керівника.

````python
# Дані — не змінюй цю клітинку
np.random.seed(2024)
base = np.random.randint(200, 800, 56)

# Вихідні продаються краще
weekend_boost = np.array([1 if i % 7 in [5, 6] else 0 for i in range(56)])
sales_56 = (base + weekend_boost * np.random.randint(100, 300, 56)).astype(int)

print('Дані отримано. Масив sales_56 готовий до аналізу.')
print(f'Елементів: {sales_56.size}, форма: {sales_56.shape}')
````

### Запитання 1 — Загальна статистика (Розділ 1, 3)
Виведи:
- Загальна виручка за 56 днів
- Середньоденна виручка (округли до 2 знаків)
- Мінімальний та максимальний денний продаж
- Стандартне відхилення (наскільки продажі «скачуть»)

````python
# Твій код:
# Загальна виручка за 56 днів
print("Загальна виручка за 56 днів: ", sales_56.sum().round(2), "грн")

# Середньоденна виручка (округли до 2 знаків)
print("Середньоденна виручка:       ", sales_56.mean().round(2), "грн")

# Мінімальний та максимальний денний продаж
print("Мінімальний денний продаж:   ", sales_56.min().round(2), "грн")
print("Максимальний денний продаж:  ", sales_56.max().round(2), "грн")

# Стандартне відхилення (наскільки продажі «скачуть»)
sales_std = np.std(sales_56)
print("Стандартне відхилення:       ", sales_std.round())
````

### Запитання 2 — Тижневий аналіз (Розділ 5)
1. Перетвори `sales_56` у матрицю **8×7** (8 тижнів по 7 днів)
2. Знайди виручку кожного тижня (сума по рядках)
3. Знайди найкращий тиждень і виведи його номер (рахуючи з 1) та суму
4. Знайди середньоденну виручку по кожному дню тижня (середнє по стовпцях)
5. Виведи який день тижня в середньому найприбутковіший (0=Пн, 6=Нд)

````python
# Твій код:
# Перетвори sales_56 у матрицю 8×7 (8 тижнів по 7 днів)
matrix_8x7 = sales_56.reshape(8, 7)
print("Matrix n8×7:\n", matrix_8x7, "\n")

# Знайди виручку кожного тижня (сума по рядках)
print("Виручка кожного тижня:", matrix_8x7.sum(axis=1))

# Знайди найкращий тиждень і виведи його номер (рахуючи з 1) та суму
print("Найкращий тиждень: №", np.argmax(matrix_8x7.sum(axis=1)) + 1)
print(f"Сума продажів за цей тиждень:", np.max(matrix_8x7.sum(axis=1)))

# Знайди середньоденну виручку по кожному дню тижня (середнє по стовпцях)
print("Середня виручка по днях тижня:", matrix_8x7.mean(axis=0))

# Виведи який день тижня в середньому найприбутковіший (0=Пн, 6=Нд)
print("Найприбутковіший день тижня:", np.argmax(matrix_8x7.mean(axis=0)))
````

### Запитання 3 — Фільтрація та маскування (Розділ 4)
1. Знайди кількість **«слабких» днів** — продажі нижче 400 грн
2. Знайди кількість **«сильних» днів** — продажі вище 700 грн
3. Яка середня виручка у слабкі дні і яка у сильні?
4. Створи масив `labels` за допомогою `np.where()`:
   - `'Слабкий'` якщо < 400
   - `'Сильний'` якщо > 700
   - `'Нормальний'` в решті випадків (підказка: вкладений `np.where`)
5. Виведи перші 14 днів: продажі та їх мітки

````python
# Твій код:
# Знайди кількість «слабких» днів — продажі нижче 400 грн
weak_days = np.sum(sales_56 < 400)
print("Кількість «слабких» днів:", weak_days)

# Знайди кількість «сильних» днів — продажі вище 700 грн
strong_days = np.sum(sales_56 > 700)
print("Кількість «сильних» днів:", strong_days, "\n")

# Яка середня виручка у слабкі дні і яка у сильні?
print("Середня виручка у слабкі дні:", sales_56[sales_56 < 400].mean().round(2))
print("Середня виручка у сильні дні:", sales_56[sales_56 > 700].mean().round(2), "\n")

# Створи масив labels за допомогою np.where():
# 'Слабкий' якщо < 400
# 'Сильний' якщо > 700
# 'Нормальний' в решті випадків (підказка: вкладений np.where)
labels = np.where(sales_56 < 400, 'Слабкий',
         np.where(sales_56 > 700, 'Сильний', 'Нормальний'))

print("labels:\n", labels, "\n")

# Виведи перші 14 днів: продажі та їх мітки
for i in range(14):
    print(f"День {i + 1}: Продажі = {sales_56[i]} -> {labels[i]}")





````
